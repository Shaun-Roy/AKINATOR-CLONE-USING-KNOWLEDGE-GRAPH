```python
import json
import csv

# Step 1: Load the JSON file
with open("all.json") as f:
    data = json.load(f)

print(f"Loaded {len(data)} characters.")
```

```python
# Step 2: Extract triples
triples = []

for hero in data:
    name = hero.get("name", "Unknown")

    # Basic traits
    gender = hero.get("appearance", {}).get("gender", "Unknown")
    race = hero.get("appearance", {}).get("race", "Unknown")
    alignment = hero.get("biography", {}).get("alignment", "Unknown")
    publisher = hero.get("biography", {}).get("publisher", "Unknown")
    occupation = hero.get("work", {}).get("occupation", "Unknown")

    # Add core triples
    triples.extend([
        ("Character", name, "HAS_GENDER", "Gender", gender),
        ("Character", name, "HAS_RACE", "Race", race),
        ("Character", name, "HAS_ALIGNMENT", "Alignment", alignment),
        ("Character", name, "HAS_PUBLISHER", "Publisher", publisher),
        ("Character", name, "HAS_OCCUPATION", "Occupation", occupation),
    ])

    # Add powerstats
    powerstats = hero.get("powerstats", {})
    for stat, value in powerstats.items():
        if value is not None and str(value).lower() != "null":
            triples.append(("Character", name, f"HAS_{stat.upper()}", "Stat", str(value)))

print(f"Extracted {len(triples)} triples.")
```

```python
# Step 3: Save triples to CSV
with open("superhero_triples.csv", "w", newline='', encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["source_label", "source", "relation", "target_label", "target"])
    writer.writerows(triples)

print("Triples saved to superhero_triples.csv")
```
